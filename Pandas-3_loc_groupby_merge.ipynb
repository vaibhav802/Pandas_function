# groupby

### Pandas groupby function is used to split the data into groups based on some criteria

```python
data = pd.read_csv('matches.csv')
data.head()
```

```python
g1 = data.groupby(by = 'city')
```

```python
g1
```

```python
g1.groups
```

```python
g1 = data.groupby(['city','team1'])
```

```python
g1
```

```python
g1.groups
```

```python
list(g1)
```

```python
dict(list(g1))
```

```python
data.groupby('city').get_group('Pune')
```

```python
data.groupby('team1').get_group('Pune Warriors')
```

# merge

### Pandas merge connects columns or indexes in DataFrame based on one or more keys.

```python
data = pd.DataFrame({'ID' : [1,2,3,4],
                   'Class' : [9,10,11,12]})
data
```

```python
data1 = pd.DataFrame({'ID' : [1,2,3,4],
                   'Name' : ['A','B','C','D']})
data1
```

```python
data2 = pd.DataFrame({'ID' : [1,2,3,5],
                   'Name' : ['A','B','C','D']})
data2
```

```python
pd.merge(data, data1, on  = 'ID')
```

```python
pd.merge(data2, data, on  = 'ID')
```

```python
pd.merge(data2, data, on  = 'ID',how  = 'inner')
```

```python
pd.merge(data2, data, on  = 'ID',how = 'left')
```

```python
pd.merge(data2, data, on  = 'ID',how = 'right')
```

```python
pd.merge(data2, data, on  = 'ID',how = 'outer')
```

```python
pd.merge(data2, data, on  = 'ID',how = 'outer', indicator = True)
```

```python
data2 = pd.DataFrame({'ID' : [5,6,7,8],
                   'Name' : ['A','B','C','D']})
data2
```

```python
data
```

```python
pd.merge(data, data2, left_index= True, right_index= True)
```

```python
data3 = pd.DataFrame({'ID' : [5,6,7,8],
                   'Name' : ['A','B','C','D']})
data3
```

```python
pd.merge(data2, data3, on = 'ID', suffixes= ('_data2','_data3'))
```
