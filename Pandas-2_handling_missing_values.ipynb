# Handling missing values


```python
import pandas as pd
```

```python
data = pd.read_csv('matches.csv')
data.head()
```

# na values

### to make some type of values as null like not available, no values

```python
data = pd.read_csv('matches.csv', na_values = ['not available','no values']) 
data.head()
```

### to make specific column nan

```python
data = pd.read_csv('matches.csv', na_values = {'team1' : ['no values','not available']})
data.head()
```

# keep_default_na

### keep nan values same like #nan as #nan not as nan

```python
data = pd.read_csv('matches.csv', keep_default_na= False) 
data.head()
```

# na_filter

### to boast performance

```python
data = pd.read_csv('matches.csv', na_filter = False) 
data.head()
```
