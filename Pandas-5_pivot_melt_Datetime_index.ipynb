# Pivot Table

###  The levels in the Pivot table will be stored in multiIndex objects( hierarchical indexes) on the index and columns of the result DataFrame

```python
import pandas as pd
```

```python
data = pd.read_csv('matches.csv')
```

```python
data
```

```python
data.pivot_table(index = 'season')
```

```python
data.pivot_table(index = 'season', columns = 'city')
```

```python
data.pivot_table(index = 'season', columns = 'city', aggfunc='count')
```

```python
data.pivot_table(index = 'season', columns = 'city', aggfunc='sum', fill_value='none')
```

```python
data.pivot_table(index = 'season', columns = 'city', aggfunc='sum', fill_value='none',margins=True)
```

# melt

### Pandas melt function is used to transform or reshape data

```python
pd.melt(data)
```

```python
pd.melt(data, id_vars = 'city')
```

```python
pd.melt(data, id_vars = ['city','id'])
```

```python
pd.melt(data, id_vars = ['city'], value_vars= ['team1'])
```

```python
pd.melt(data, id_vars = ['city'], value_vars= ['team1'], var_name='Category')
```

```python
pd.melt(data, id_vars = ['city'], value_vars= ['team1'], var_name='Category', value_name='Team1')
```
